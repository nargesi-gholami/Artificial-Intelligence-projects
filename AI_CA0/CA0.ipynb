```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

part1


```python
data = pd.read_csv('FuelConsumptionCo2.csv')
fuelData = pd.DataFrame(data)
print(fuelData.head())
print(fuelData.describe())
print(fuelData.tail())
```

part2

```python
fuelData.info(verbose=True)

fuelData['FUELTYPE'] = fuelData['FUELTYPE'].astype('category')
fuelData['FUELTYPE'] = fuelData['FUELTYPE'].cat.codes
fuelData.tail(n = 50)
```

part3

```python
count_nan = len(fuelData) - fuelData.count()
print(count_nan)
```

```python
numeric_columns = fuelData.select_dtypes(include=['number']).columns.difference(['CO2EMISSIONS'])
fuelData[numeric_columns] = fuelData[numeric_columns].fillna(fuelData.mean(numeric_only=True))

Categorical_columns = fuelData.select_dtypes(exclude=['number']).columns
fuelData[Categorical_columns] = fuelData[Categorical_columns].transform(lambda a: a.fillna(a.mode()[0]))

count_nan2 = len(fuelData) - fuelData.count()
print(count_nan2)

NanNumber = fuelData['CO2EMISSIONS'].isnull()
NotNanNumber = ~fuelData['CO2EMISSIONS'].isnull()
fuelDataNan = fuelData[NanNumber]
fuelDataNotNan = fuelData[NotNanNumber]
fuelDataNotNan = fuelDataNotNan.reset_index()
print(fuelDataNan)
print(fuelDataNotNan)
```

part4

```python
startTime1 = time.time()
goal_columns2 = fuelDataNotNan['CO2EMISSIONS'] < 240
fuelData2 = fuelDataNotNan.groupby(goal_columns2)
meanFuel2 = fuelData2['FUELCONSUMPTION_CITY'].mean()
print(meanFuel2[1])

goal_columns3 = fuelDataNotNan['CO2EMISSIONS'] > 300
fuelData3 = fuelDataNotNan.groupby(goal_columns3)
meanFuel3 = fuelData3['FUELCONSUMPTION_CITY'].mean()
print(meanFuel3[1])
endTime1 = time.time()
```

```python
startTime2 = time.time()
count1 = 0
sum1 = 0
for i in range(len(fuelDataNotNan)):
    if fuelDataNotNan['CO2EMISSIONS'][i] < 240:
        sum1 += fuelDataNotNan['FUELCONSUMPTION_CITY'][i]
        count1 += 1       
print(sum1 / count1)

count2 = 0
sum2 = 0
for i in range(len(fuelDataNotNan)):
    if fuelDataNotNan['CO2EMISSIONS'][i] > 300:
        sum2 += fuelDataNotNan['FUELCONSUMPTION_CITY'][i]
        count2 += 1
print(sum2 / count2)
endTime2 = time.time()
```

part5

```python
print(endTime1 - startTime1)
print(endTime2 - startTime2)
```

part6

```python
fuelDataNotNan.hist(bins=25, grid=False, figsize=(15,12), color='#86bf92', zorder=2, rwidth=0.9)

numericFuelData = fuelDataNotNan.select_dtypes(include=['number']).columns.difference(['FUELTYPE'])
fuelDataNotNan2 = fuelDataNotNan[numericFuelData]
print(fuelDataNotNan2)
```

part7

```python
newNumericFuelData = (fuelDataNotNan2 - fuelDataNotNan2.mean()) / fuelDataNotNan2.std()
newNumericFuelData
```

part 9

```python
fuelDataNew = newNumericFuelData[['CO2EMISSIONS', 'FUELCONSUMPTION_COMB']]
```

part 10

```python
A = np.arange(1*2.00).reshape(2, 1)
X = np.arange(len(fuelDataNew)*2.00).reshape(len(fuelDataNew), 2)
X[:, 0] =  1
X[:, 1] = fuelDataNew['FUELCONSUMPTION_COMB']
Y = fuelDataNew['CO2EMISSIONS']
X_pinv = np.linalg.pinv(X)
A = np.dot(X_pinv,Y)
A
```

```python
H = A[1]*fuelDataNew['FUELCONSUMPTION_COMB'] + A[0]
MSE = sum((fuelDataNew['CO2EMISSIONS'] - H)**2)/len(fuelDataNew)
print(MSE)
```
